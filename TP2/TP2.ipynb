### 1. ¿Qué motivación tenemos para utilizar algoritmos basados en la inferencia lógica en lugar de utilizar algoritmos de búsqueda para cualquier tipo de problema?

Principalmente: el menor uso computacional que requieren los algoritmos basados en la inferencia lógica.

Los algoritmos de búsqueda requieren tener cargado en memoria todas los posibles estados, o una regla lógica para cada una, y recorrer todas las posibilidades. Debido a que los estados pueden adoptar valores continuo, el problema de busqueda se expande exponencialmente. 

Además, los algoritmos de búsqueda no se adaptan a cambios del entorno, mientras que los algoritmos basados en inferencia lógica se deducen de la misma base de conocimiento, por lo que son adaptables a cualquier cambio.

Por otro lado, los algoritmos de inferencia lógica son mas "complicados" de plantear en un principio, ya que debe definirse de forma correcta la base de conocimiento; una vez que ésta está bien planteada, es mucho más fácil que se adapte a distintos problemas.

Se puede concluir entonces que: los algoritmos de lógica difusa son **más robustos** frente a errores de diseño pero son **menos efectivos**.



### 2. ¿Qué diferencia hay entre una base de conocimientos y el ground truth?

Es útil definir ambos conceptos antes de diferenciarlos.

**Base de Conocimeinto**

> La Base de Conocimiento consiste en el conjunto de reglas, axiomas y relaciones que definene el contexto del problema. Se basa en reglas o teorías estudiadas. 

**Ground Truth**

> Corresponde a los hechos que se dan en una instancia inicial del dominio. Corresponde a los valores sensados en un inicio que ayudan a complementar las reglas. 
>
> Por ejemplo, el valor de la aceleración es un valor sensado pero preevisto de antemano para la resolución del problema, por lo que corresponde al ground truth.

El Ground Truth se puede ver como el estímulo que activa la Base de Conocimiento. Usando una analogía con Sistemas Dinámicos: 
- Base de Conocimientos $\approx$ Ecuación Diferencial: Define el comportamiento universal y las relaciones entre variables del sistema ($m\ddot{x} + c\dot{x} + kx = F$).
- Ground Truth $\approx$ Condiciones Iniciales / Entradas: Son los valores específicos ($x_0, v_0$) y las excitaciones externas que permiten resolver la ecuación para un caso concreto.

> - Origen: Surge del Motor de Inferencia y pasa por un proceso de Desborrosificación para volver al entorno.

> - Ejemplo: Un ciclo de trabajo PWM del $60 \%$ para un ventilador, inferido a partir de la temperatura percibida.




### 4. ¿Qué valores puede tomar una oración atómica de acuerdo a la lógica proposicional? ¿Cómo se modela la frase “Los valores de A comprendidos entre 0 y 9” de acuerdo a esta sintaxis?

La lógica proposicional es una lógica binaria que admite solo dos valores **discretos** y **mutuamente excluyentes**: sus variables pueden tomar solo valores **verdadero** (1) o **falso** (0). 

Debo definir como como proposiciones la existencia de cada número: por ejemplo $A_0$ significa "<i>el número 0 existe/esta presente</i>".

Tiene una gran cantidad de limitaciones. 

**Incapacidad de manejar Grados de Verdad**
> No permite representar situaciones intermedias o verdades parciales.

**Ineficiencia en Sistemas Continuos (o Sistemas Discretos Masivos)**
> La lógica binaria está diseñada para dominios discretos. Intentar modelar una variable continua (como la aceleración) mediante proposiciones individuales ($A_0, A_1, \dots, A_n$) requiere una cantidad masiva de axiomas, lo que vuelve al sistema ineficiente y difícil de mantener.Rigidez en 

**Inestabilidad en la Respuesta**
> Para sistemas que operan en tiempo real, la falta de una transición suave entre estados (característica de la lógica multi-valuada) genera respuestas de control bruscas o inestables.

La frase "Los valores de A comprendidos entre 0 y 9" puede modelarse como:

$$
S = A_{0} \lor A_{1} \lor A_{2} \dots \lor A_{9}
$$

$$
\begin{array}{ll}
  \textbf{Donde:} & \\
  \qquad A_{k} & \text{El número k esta presente}  \\
\qquad S & \text{Valor de verdad de la sentencia.}
\end{array}
$$


### 5. ¿Qué algoritmos se utilizan para resolver problemas modelados con lógica proposicional?

En los sistemas basados en lógica preposicional, la obtención de nuevas conclusiones a partir de una base de conocimientos y una base de hechos (Ground Truth) se realiza mediante **motores de inferencia** que utilizan principalmente tres estrategias:

**A. Encadenamiento hacia adelante (Forward Chaining)**

> Es un algoritmo **dirigido por los datos** (_data-driven_).
> 
> **Funcionamiento:** Parte de los hechos conocidos (Ground Truth) y aplica las reglas de la base de conocimientos para inferir nuevos hechos. Este proceso se repite de forma iterativa hasta que se alcanza el objetivo o no se pueden disparar más reglas.
>     

**B. Encadenamiento hacia atrás (Backward Chaining)**

> Es un algoritmo **dirigido por los objetivos** (_goal-driven_).
> 
>**Funcionamiento:** Parte de una hipótesis o meta que se desea probar. El motor busca en la base de conocimientos aquellas reglas que tengan dicha meta como consecuente y luego intenta verificar si los antecedentes de esas reglas son ciertos (buscándolos en la base de hechos o probándolos como sub-metas).
>     

 **C. Demostración por Contradicción (Resolución)**

> Es un método de refutación lógica.
> 
> - **Funcionamiento:** Para demostrar que una proposición $P$ es verdadera, el algoritmo asume que $\neg P$ (la negación de $P$) es verdadera y busca generar una inconsistencia o contradicción dentro de la base de conocimientos. Si se halla la contradicción, por reducción al absurdo, $P$ debe ser verdadera.
>     


### 6. ¿Qué valores puede tomar una oración atómica de acuerdo a la lógica de primer orden? ¿Cómo se modela la frase “Los valores de A comprendidos entre 0 y 9” de acuerdo a esta sintaxis

La Lógica de Primer Orden (LPO), además de las herramientas de la lógica preposicional, permite **utilizar cuantificar** sobre objetos y **definir relaciones**, pero aún sigue limitada por valores de verdad binarios.

Se debe tener en mente que:

* Al igual que la proposicional, la LPO falla al intentar modelar la verdad parcial. No puede representar que un motor está "un poco caliente"; para la LPO, o está en el conjunto de las cosas calientes (1) o no lo está (0).

* Sigue un enfoque top-down, donde el conocimiento se inyecta mediante reglas rígidas definidas por el programador.


La frase "Los valores de A comprendidos entre 0 y 9" puede modelarse como:
$$
S = (0\leq A) \land (A\leq9)
$$
ó
$$
S= \exists A \in [0,9]
$$

$$
\begin{array}{ll}
  \textbf{Donde:} & \\
  \qquad A & \text{Es la variable (es un número }\mathbb{R}\text{)}  \\
\qquad S & \text{Valor de verdad de la sentencia.}
\end{array}
$$



### 7. ¿Qué algoritmos se utilizan para resolver problemas modelados con lógica de primer orden?

A diferencia de la lógica proposicional, la LPO requiere mecanismos capaces de manejar variables y cuantificadores ($\forall, \exists$). El proceso de resolución se fundamenta en la búsqueda dentro de un espacio de estados donde cada nodo representa una combinación lógica de predicados

Para ello, se utilizan Lenguajes que hacen uso de motores de inferencia. **Todos los motores de inferencia se basan en algoritmos de búsqueda**, donde cada uno de los estados es una combinación lógica de las variables. Lo bueno que tienen los motores de inferencia es que se pueden configurar en base a si queremos una **solución más rápida** o **más precisa**.

Algunos lenguajes de programación que hacen uso de motores de inferencia con LPO es PROLOG
> PROLOG (PROgrammation en LOGique) opera mediante un motor de inferencia con un enfoque top-down que utiliza la lógica simbólica de primer orden para deducir conclusiones a partir de una base de conocimientos conformada por hechos y reglas. PROLOG emplea de forma nativa una estrategia de búsqueda primero en profundidad (DFS) combinada con backtracking y unificación para navegar el árbol de objetivos hasta validar una consulta.

- Utilice el siguiente modelo del sistema carro-péndulo

Modelado del sistema físico:


Haciendo uso del método del operador Lagrangiano, desarrollamos las ecuaciones de movimiento:
 $$
 \begin{align*}
 T_{M}  & = \frac{1}{2} m_{c} \dot{x}^{2} \\
 V_{M}  & = 0 \\
 D_{M}  & = ? \\ \\

 T_{m}  & = \frac{1}{2} m_{p} (\dot{x}_{p}^{2}+ \dot{y}_{p}^{2}) + \frac{1}{2} J_{G} \dot{\theta}^{2}\\
 V_{m}  & = \frac{1}{2} m_{p}gl \cos \theta \\
 D_{m}  & = ?
 \end{align*}
$$
Consideramos coeficientes de fricción viscosa nulos, por lo que la función de viscosidad de Raylight queda:

$$
D_{M} = D_{m} = 0
$$

Analizando más minuciosamente la energía cinética de la masa de péndulo, vemos que aparecen los términos $\dot{x}_{p}$ y $\dot{y}_{p}$. Esto sucede debido a que el péndulo sufre de movimiento plano (movimiento angular + movimiento linear).

Si se condieran los reemplazos:
$$
\begin{array}{ll}
  \textbf{Reemplazando:} & \\
  \qquad x_{p}=x+l\sin(\theta) &  \\
 \qquad y_{p}=l\cos (\theta) &   \\
 \qquad v^{2}_{p}=\dot{x}^{2}_{p} + \dot{y}^{2}_{p} &
\end{array}
$$

Por relaciones trigonométricas:
 $$
 \begin{align*}
 \dot{x}_{p}^{2} + \dot{y}_{p}^{2}  & = \left( \frac{d}{dt} \left( x+l \sin (\theta) \right)  \right)^{2} +\left( \frac{d}{dt} l \cos (\theta)  \right)^{2}  \\
  \\
 & = l^{2} \dot{\theta}^{2} (\cos^{2}(\theta) + \sin ^{2}(\theta)) + \dot{x} ^{2} + 2l\dot{x} \dot{\theta}\cos(\theta) 
 \\
 & = l^{2} \dot{\theta}^{2} + \dot{x}^{2} + 2l \dot{x} \dot{\theta}\cos(\theta)
 \end{align*}
$$

lo que nos permite dejar:

$$
T_{m} = \frac{1}{2}(J_{g}+m_{p}l^{2})\dot{\theta}^{2} + \frac{1}{2} m_{p} \dot{x}^{2} + m_{p}l \dot{x} \dot{\theta} \cos(\theta)
$$

$$
\begin{array}{ll}
  \textbf{Considerando:} & \\
  \qquad J_{G} = \frac{1}{3}m_{p}l^{2} &  \\
 \qquad \lambda= J_{O} = \frac{4}{3}m_{p}l^{2} &   \\
 \qquad \alpha=m_{c}+m_{p} &   \\
 \qquad \mu = m_{p}l &  
\end{array}
$$


Y llegamos a:
 $$
\begin{align*}
 T  & = \frac{1}{2} \alpha \dot{x}^{2} + \frac{1}{2} \lambda \dot{\theta}^{2} +\mu \dot{x} \dot{\theta}\cos(\theta) \\
 V & = \mu g \cos \theta \\
 D  & = 0
\end{align*}
$$
 Reemplazando para cada coordenada generalizada en la expresión:

 $$
 \frac{d}{dt} \Big( \frac{\partial T}{\partial \dot{q_k}} \Big) - \frac{\partial T}{\partial q_k} + \frac{\partial V}{\partial{q_k}} +\frac{\partial D_v}{\partial \dot{q_k}} = Q_k
$$
 
 Con esto, llegamos a:

$$
\begin{align*} 
\alpha \ddot{x} + \mu \Big( \ddot{\theta}\cos(\theta) - \dot{\theta}^{2}\sin(\theta ) \Big) &  = F \\
\lambda \ddot{\theta} + \mu \ddot{x} \cos(\theta)  - \mu g \sin(\theta)  & = M
\end{align*}
$$


> Se tuvo:
> $$
> \lambda \ddot{\theta} + \Big(\mu \ddot{x} \cos(\theta) -\cancel{  \mu \dot{x} \dot{\theta}\sin(\theta) }\Big) +\cancel{ \mu \dot{x}\dot{\theta}\sin(\theta) } - \mu g \sin(\theta) = M
> $$

Recordamos que $F$ es la fuerza de acción y $M$ el torque aplicado al péndulo. 

Si bien la posición del carrito, denotada por la coordenada generalizada $x$, es importante, lo que nos interesa en principio es que el péndulo quede vertical. Por ello, por ahora asumiremos que el carrito no tiene restricción de movimiento y con ello despejamos:
$$
\ddot{x}   = \frac{F - \mu \Big( \ddot{\theta}\cos(\theta) - \dot{\theta}^{2}\sin(\theta ) \Big)}{\alpha }
$$
Debido a que solo aparece $\ddot{x}$ en la ecuación de movimiento de $\theta$ se puede reemplazar esta expresión y despejar $\ddot{\theta}$. 

$$
\begin{align*}
\lambda \ddot{\theta} + \mu \ddot{x} \cos(\theta)  - \mu g \sin(\theta)  & = M \\
 \\
\lambda \ddot{\theta} + \mu \cos(\theta) \Bigg( \frac{F - \mu \Big( \ddot{\theta}\cos(\theta) - \dot{\theta}^{2}\sin(\theta ) \Big)}{\alpha } \Bigg)  - \mu g \sin(\theta)  & = M \\
 \\
\Big( \lambda - \frac{\mu^{2}\cos ^{2}(\theta)}{\alpha} \Big) \ddot{\theta} + \mu \cos(\theta) \Bigg( \frac{F + \mu \dot{\theta}^{2}\sin(\theta) }{\alpha } \Bigg)  - \mu g \sin(\theta)  & = M 
\end{align*}
$$
Despejando $\ddot{\theta}$:
$$
\ddot{\theta} = \frac{M + \mu g \sin(\theta)  - \mu \cos(\theta) \Bigg( \frac{F + \mu \dot{\theta}^{2}\sin(\theta) }{\alpha } \Bigg)}{ \lambda - \frac{\mu^{2}\cos ^{2}(\theta)}{\alpha} }
$$

$$
\begin{align*}
\ddot{\theta} = \frac{\cancel{ M } + \mu g \sin(\theta)  - \mu \cos(\theta) \Bigg( \frac{F + \mu \dot{\theta}^{2}\sin(\theta) }{\alpha } \Bigg)}{ \lambda - \frac{\mu^{2}\cos ^{2}(\theta)}{\alpha} } \\

\end{align*}
$$
Esto nos deja con la misma ecuación que la consigna, solo se deben reemplazar los términos por las unidades físicas (reemplazar $\alpha$, $\lambda$ y $\mu$).


> Igualamos la nomenclatura:
