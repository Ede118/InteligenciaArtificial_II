1. ¿Qué motivación tenemos para utilizar algoritmos basados en la inferencia lógica en lugar de utilizar algoritmos de búsqueda para cualquier tipo de problema?

Menor uso computacional.
Para la busqueda se necesita tener cargado en memoria todas las posibilidades o una regla lógica para cada uno y se necesita recorrer todas las posibilidades. Al ser un valor continuo el que se busca se puede expandir muchísimo la búsqueda. 
Los algoritmos de búsqueda no se adaptan a cambios del entorno, mientras que los algoritmos basados en inferencia lógica se deducen de la misma base de conocimiento, por lo que son adaptables a cualquier cambio.
Por otro lado, los algoritmos de inferencia lógica son mas "complicados" de plantear en un principio, ya que debe definirse de forma correcta la base de conocimiento, pero una vez que ésta está bien planteada, es mucho más fácil que se adapte a distintos problemas.

Loas algoritmos de lógica difusa son más robustos frente a errores de diseño, pero son menos efectivos.

2. ¿Qué diferencia hay entre una base de conocimientos y el ground truth?

Una base de conocimeinto consiste en el conjunto de reglas, axiomas y relaciones que definene el contexto del problema. Se basa en reglas o teorías estudiadas. Mientras que el ground truth corresponde a los hechos que se dan en una instancia inicial del dominio. Corresponde a los valores sensados en un inicio que ayudan a complementar las reglas.
Por ejemplo, el valor de la aceleración es un valor sensado pero preevisto de antemano para la resolución del problema, por lo que corresponde al ground truth.

3. ¿Qué diferencia hay entre un valor inferido y un valor percibido?

El valor percibido proviene de un valor sensado o conocido de antemano, mientras que un valor inferido se calcula o deriva de una regla que combina valores percibidos.
